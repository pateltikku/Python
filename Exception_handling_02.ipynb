Q5. Explain ImportError. What is ModuleNotFoundError?

1.ImportError- When there is a problem while trying to import a module, ImportError takes place. This can happen for various reasons, such as when the module is not found, there are syntax errors in the module, or there are issues with the module's dependencies. A relevant example is demonstrated below-

```python
try:
    import non_existent_module
except ImportError as e:
    print(f"ImportError: {e}")
```

2. ModuleNotFoundError- It occurs when python interpreter is unable to locate the module, for which import has been executed. A relevant example is demonstrated below-

```python
try:
    import non_existent_module
except ModuleNotFoundError as e:
    print(f"ModuleNotFoundError: {e}")
```

Q6. List down some best practices for exception handling in python. 


1. Catch specific exception types rather than using a broad except clause. This helps you handle errors more accurately and avoids catching unexpected exceptions.

2. If the need arises of handling of different exceptions differently, employing multiple except blocks as a solution proves  to be useful.

3. The finally block is used to define code that should be executed regardless of whether an exception was raised or not. It's often used for cleanup tasks like closing files or releasing resources.

4. Using the else block with a try statement to specify code that should run only if no exceptions were raised. This is useful for separating the regular code path from the exception handling.

5. For more complex applications, defining custom exception classes is one of the effective methods that provide meaningful information about the error with the context.

6. Using Python's built-in logging module to log exceptions is a great practice which helps in tracking and diagnosing issues in production environments.
